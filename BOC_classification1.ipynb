```python
import numpy as np
import pandas as pd
import random

random.seed(7)
```

```python
df20 = pd.read_csv('/content/feature20.csv')
df21 = pd.read_csv('/content/feature21.csv')
```

```python
df20 = df20.drop(['Week_Number'], axis=1)
df21 = df21.drop(['Week_Number'], axis=1)
```

```python
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Split the data into features (X) and target (y)
X = df20.drop('z20', axis=1)
y = df20['z20']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Scale the features using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#k=3
from sklearn.metrics import accuracy_score
model1 = KNeighborsClassifier(n_neighbors=3)
model1.fit(X_train, y_train)

y_pred = model1.predict(X_test)
accuracy1 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy1)
```

```python
#k=5
model2 = KNeighborsClassifier(n_neighbors=5)
model2.fit(X_train, y_train)

y_pred = model2.predict(X_test)
accuracy2 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy2)
```

```python
#k=7
model3 = KNeighborsClassifier(n_neighbors=7)
model3.fit(X_train, y_train)

y_pred = model3.predict(X_test)
accuracy3 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy3)
```

```python
#k=9
model4 = KNeighborsClassifier(n_neighbors=9)
model4.fit(X_train, y_train)

y_pred = model4.predict(X_test)
accuracy4 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy4)
```

```python
#k=11
model5 = KNeighborsClassifier(n_neighbors=11)
model5.fit(X_train, y_train)

y_pred = model5.predict(X_test)
accuracy5 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy5)
```

```python
#best model : model1
X1 = df21.drop('z21', axis=1)
y1 = df21['z21']
X1 = scaler.fit_transform(X1)

y_pred = model1.predict(X1)
acc = accuracy_score(y1, y_pred)
```
